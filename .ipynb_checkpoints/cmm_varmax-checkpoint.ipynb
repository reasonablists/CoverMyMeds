```python
import numpy as np
import pandas as pd
import pyarrow
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import timedelta
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARMA
from statsmodels.tsa.statespace.varmax import VARMAX
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
from time import time
```

# Fitting a VARMAX Model

```python
# Import the data and set the index to dates
cmm = pd.read_parquet('cmm_erdos_bootcamp_2020_timeseries.pq', engine='pyarrow')
cmm.set_index('date_val', inplace=True)
# Provide frequency information for the model to avoid error messages
cmm = cmm.asfreq(pd.infer_freq(cmm.index))
```

```python
# Volumes must be cast as type float to avoid error messages in VARMAX
cmm['volume_A'] = cmm['volume_A'].astype(float)
cmm['volume_B'] = cmm['volume_B'].astype(float)
cmm['volume_C'] = cmm['volume_C'].astype(float)
```

```python
# Specifiy a start date and end date of the entire data
start_date = pd.to_datetime('2018-01-01')
end_date = pd.to_datetime('2019-12-31')
cmm = cmm[start_date:end_date]
```

```python
# Declare endogenous variables and arrange them in a vector
cmm_va = cmm['volume_A']
cmm_vb = cmm['volume_B']
cmm_vc = cmm['volume_C']
cmm_vec = pd.DataFrame({'vol_A': cmm_va, 'vol_B': cmm_vb, 'vol_C': cmm_vc})
```

```python
# Declare possible exogenous variables and arrange them in a vector
cmm_exog = pd.DataFrame({'calendar_year': cmm['calendar_year'], 
                         'calendar_month': cmm['calendar_month'], 
                         'calendar_day': cmm['calendar_day'], 
                         'day_of_week': cmm['day_of_week'], 
                         'is_weekday': cmm['is_weekday'], 
                         'is_workday': cmm['is_workday'], 
                         'is_holiday': cmm['is_holiday']})
```

# Predictions without a Rolling Forecast

```python
# Declare training data and test data
train_end = pd.to_datetime('2019-9-1')
test_end = pd.to_datetime('2019-10-1')
train_data = cmm_vec[:train_end]
test_data = cmm_vec[train_end + timedelta(days=1):test_end]
```

```python
# Fit the VARMAX model and print a summary
model = VARMAX(endog=train_data, exog=cmm_exog[:train_end], order=(7,0))
model_fit = model.fit()
print(model_fit.summary())
```

```python
# Make predictions based on model and define residuals
pred_start_date = test_data.index[0]
pred_end_date = test_data.index[-1]
predictions = model_fit.predict(start=pred_start_date, end=pred_end_date, exog=cmm_exog[pred_start_date:pred_end_date])
residuals_va = test_data['vol_A'] - predictions['vol_A']
residuals_vb = test_data['vol_B'] - predictions['vol_B']
residuals_vc = test_data['vol_C'] - predictions['vol_C']
```

```python
# Compute the MAPE and RMSE for each time series using VARIMAX
print('Mean Absolute Percent Error for A:', round(np.mean(abs(residuals_va/test_data['vol_A'])),4))
print('RMS Error for A:', np.sqrt(np.mean(residuals_va**2)))
print('Mean Absolute Percent Error for B:', round(np.mean(abs(residuals_vb/test_data['vol_B'])),4))
print('RMS Error for B:', np.sqrt(np.mean(residuals_vb**2)))
print('Mean Absolute Percent Error for C:', round(np.mean(abs(residuals_vc/test_data['vol_C'])),4))
print('RMS Error for C:', np.sqrt(np.mean(residuals_vc**2)))
```

```python
# Plot residuals
plt.figure(figsize=(10,4))
plt.plot(residuals_va)
plt.plot(residuals_vb)
plt.plot(residuals_vc)
plt.legend(('Residual A', 'Residual B', 'Residual C'), fontsize=16)
plt.title('Residuals from VARMAX Model', fontsize=20)
plt.ylabel('Error', fontsize=16)
plt.axhline(0, color='r', linestyle='--', alpha=0.2)
plt.show()
```

```python
# Plot predictions
plt.figure(figsize=(10,4))
plt.plot(test_data['vol_A'])
plt.plot(predictions['vol_A'])
plt.legend(('Data', 'Predictions'), fontsize=16)
plt.title('Volume A over Time', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(test_data['vol_B'])
plt.plot(predictions['vol_B'])
plt.legend(('Data', 'Predictions'), fontsize=16)
plt.title('Volume B over Time', fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.plot(test_data['vol_C'])
plt.plot(predictions['vol_C'])
plt.legend(('Data', 'Predictions'), fontsize=16)
plt.title('Volume C over Time', fontsize=20)
plt.show()
```
