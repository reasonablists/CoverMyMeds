```python
#Importing all packages

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
#Package for the parquet file
import pyarrow.parquet as pq
```

```python
df_cmm = pq.read_table(r'C:\Users\k2sin\Erdos_Inst_Bootcamp-2020\Code2020Materials\cmm_erdos_bootcamp_2020_timeseries.pq').to_pandas()
```

```python
df_cmm.head()
```

```python
#Convering date_val to datetime format

df_cmm_mod = df_cmm
df_cmm_mod['date_val'] = pd.to_datetime(df_cmm_mod['date_val'])
```

```python
#Adding a column depicting the day of the week

days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
day_of_week = []
for i in range(len(df_cmm_mod)):
    day_of_week.append(days[i%7])
    
df_cmm_mod['Day_of_week'] = day_of_week
df_cmm_mod.head()
```

# Monthly plots for 2017

```python
# Array depicting indiced for the months

months_arr = [0,31,59,90,120,151,181,212,243,273,304,334,365]
```

```python
fig, ax = plt.subplots(4,3,figsize = (16,18))
months = ['January','February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November',
          'December']
i=0
for j in range(12):
    beginning = (i+1)*months_arr[j]
    end = (i+1)*months_arr[j+1]
    month = df_cmm_mod.iloc[beginning:end]
    ax[j//3,j%3].plot(range(1,end - beginning+1),month['volume_A'], label = 'volume_A')
    ax[j//3,j%3].plot(range(1,end - beginning+1),month['volume_B'], label = 'volume_B')
    ax[j//3,j%3].plot(range(1,end - beginning+1),month['volume_C'], label = 'volume_C')
    title = months[j]
    ax[j//3,j%3].set_title(title + ' 2017')
    ax[j//3,j%3].grid()
    ax[j//3,j%3].set_xticks(range(1,end-beginning+1))
    ax[j//3,j%3].set_xticklabels(df_cmm_mod['Day_of_week'].iloc[beginning:end],rotation=90,fontsize = 7.5)
handles,labels = ax[0,0].get_legend_handles_labels()
fig.legend(handles,labels,loc='upper center')
```

# Mean and total volume plots for all years

```python
## Year 2017
mean_A_17 = []
mean_B_17 = []
mean_C_17 = []
for j in range(12):
    beginning = months_arr[j]
    end = months_arr[j+1]
    a = df_cmm_mod.iloc[beginning:end]['volume_A'].mean()
    b = df_cmm_mod.iloc[beginning:end]['volume_B'].mean()
    c = df_cmm_mod.iloc[beginning:end]['volume_C'].mean()
    mean_A_17.append(a)
    mean_B_17.append(b)
    mean_C_17.append(c)         
```

```python
## Year 2018
mean_A_18 = []
mean_B_18 = []
mean_C_18 = []
for j in range(12):
    beginning = months_arr[j]
    end = months_arr[j+1]
    a = df_cmm_mod.iloc[beginning+365:end+365]['volume_A'].mean()
    b = df_cmm_mod.iloc[beginning+365:end+365]['volume_B'].mean()
    c = df_cmm_mod.iloc[beginning+365:end+365]['volume_C'].mean()
    mean_A_18.append(a)
    mean_B_18.append(b)
    mean_C_18.append(c)  
```

```python
## Year 2019
mean_A_19 = []
mean_B_19 = []
mean_C_19 = []
for j in range(12):
    beginning = months_arr[j]
    end = months_arr[j+1]
    a = df_cmm_mod.iloc[beginning+730:end+730]['volume_A'].mean()
    b = df_cmm_mod.iloc[beginning+730:end+730]['volume_B'].mean()
    c = df_cmm_mod.iloc[beginning+730:end+730]['volume_C'].mean()
    mean_A_19.append(a)
    mean_B_19.append(b)
    mean_C_19.append(c)  
```

```python
fig, ax = plt.subplots(3,1,figsize = (16,14))

## Volume_A plot
ax[0].plot(range(1,13),mean_A_17,'r', label = '2017')
ax[0].plot(range(1,13),mean_A_18,'b', label = '2018')
ax[0].plot(range(1,13),mean_A_19,'k', label = '2019')

ax[0].set_title('Mean Volume Plots for volume_A')
ax[0].set_xlabel('Months')
ax[0].set_ylabel('Volume')
ax[0].set_xticks(range(1,13))
ax[0].legend()
ax[0].grid()

## Volume_B plot

ax[1].plot(range(1,13),mean_B_17,'r', label = '2017')
ax[1].plot(range(1,13),mean_B_18,'b', label = '2018')
ax[1].plot(range(1,13),mean_B_19,'k', label = '2019')

ax[1].set_title('Mean Volume Plots for volume_B')
ax[1].set_xlabel('Months')
ax[1].set_ylabel('Volume')
ax[1].set_xticks(range(1,13))
ax[1].legend()
ax[1].grid()

## Volume_C plot

ax[2].plot(range(1,13),mean_C_17,'r', label = '2017')
ax[2].plot(range(1,13),mean_C_18,'b', label = '2018')
ax[2].plot(range(1,13),mean_C_19,'k', label = '2019')

ax[2].set_title('Mean Volume Plots for volume_C')
ax[2].set_xlabel('Months')
ax[2].set_ylabel('Volume')
ax[2].set_xticks(range(1,13))
ax[2].legend()
ax[2].grid()
```

```python
def agg(i):    
    sum_A = []
    sum_B = []
    sum_C = []
    for j in range(12):
        beginning = months_arr[j]
        end = months_arr[j+1]
        a = df_cmm_mod.iloc[beginning+365*i:end+365*i]['volume_A'].sum()
        b = df_cmm_mod.iloc[beginning+365*i:end+365*i]['volume_B'].sum()
        c = df_cmm_mod.iloc[beginning+365*i:end+365*i]['volume_C'].sum()
        sum_A.append(a)
        sum_B.append(b)
        sum_C.append(c)  
    return sum_A, sum_B, sum_C
```

```python
[sum_A_17, sum_B_17, sum_C_17]= agg(0)
[sum_A_18, sum_B_18, sum_C_18]= agg(1)
[sum_A_19, sum_B_19, sum_C_19] = agg(2)
```

```python
fig, ax = plt.subplots(3,1,figsize = (16,14))

## Volume_A plot
ax[0].plot(range(1,13),sum_A_17,'r', label = '2017')
ax[0].plot(range(1,13),sum_A_18,'b', label = '2018')
ax[0].plot(range(1,13),sum_A_19,'k', label = '2019')

ax[0].set_title('Total Volume Plots for volume_A')
ax[0].set_xlabel('Months')
ax[0].set_ylabel('Volume')
ax[0].set_xticks(range(1,13))
ax[0].legend()
ax[0].grid()

## Volume_B plot

ax[1].plot(range(1,13),sum_B_17,'r', label = '2017')
ax[1].plot(range(1,13),sum_B_18,'b', label = '2018')
ax[1].plot(range(1,13),sum_B_19,'k', label = '2019')

ax[1].set_title('Total Volume Plots for volume_B')
ax[1].set_xlabel('Months')
ax[1].set_ylabel('Volume')
ax[1].set_xticks(range(1,13))
ax[1].legend()
ax[1].grid()

## Volume_C plot

ax[2].plot(range(1,13),sum_C_17,'r', label = '2017')
ax[2].plot(range(1,13),sum_C_18,'b', label = '2018')
ax[2].plot(range(1,13),sum_C_19,'k', label = '2019')

ax[2].set_title('Total Volume Plots for volume_C')
ax[2].set_xlabel('Months')
ax[2].set_ylabel('Volume')
ax[2].set_xticks(range(1,13))
ax[2].legend()
ax[2].grid()
```

# Observations

1. For customer A, the volume increases over the year in 2017 but shows no consistent trend for the years 2018 and 2019. However, for all three years, we see a sharp increase in volume in the months of March and October.
2. For customer B, the volume decreases as the year progresses for each of 2017, 2018 and 2019. Again, we observe a sharp increase in volume for the months of March and October.
3. For customer C, the volume decreases as the year progresses for both 2017 and 2019, while there is no consistent trend for 2018. 
4. The total volume for all three medicines increases every year from 2017 to 2019.
